```python
import pandas as pd


# One-hot encode the categorical variables
total_data_encoded = pd.get_dummies(total_data, drop_first=True)

# Now, I separate my features and target variable again
X = total_data_encoded.drop(columns=['y_yes'])  
y = total_data_encoded['y_yes']  
```

```python
from sklearn.model_selection import train_test_split

#spli the data again 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import chi2, SelectKBest

# Initialize the MinMaxScaler becuase the X ValueError: Input X must be non-negative. 
#error when applying SelectKBest with chi2 to your dataset X_train 
scaler = MinMaxScaler()

# Fit the scaler on the training data and transform both training and test sets
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Apply SelectKBest with chi2 on the scaled data
selection_model = SelectKBest(chi2, k=5)
selection_model.fit(X_train_scaled, y_train)

# Get the selected features
ix = selection_model.get_support()

# Apply the selection to both training and test sets
X_train_sel = X_train_scaled[:, ix]
X_test_sel = X_test_scaled[:, ix]
```

```python
X_test_sel
```

```python
# Convert the numpy arrays back to pandas DataFrames
X_train_sel_df = pd.DataFrame(X_train_sel, columns=X_train.columns.values[ix])
X_test_sel_df = pd.DataFrame(X_test_sel, columns=X_test.columns.values[ix])

# Add the target variable back
X_train_sel_df['y_n'] = y_train.reset_index(drop=True)
X_test_sel_df['y_n'] = y_test.reset_index(drop=True)

# Save the datasets to CSV files
X_train_sel_df.to_csv("../data/processed/clean_train.csv", index=False)
X_test_sel_df.to_csv("../data/processed/clean_test.csv", index=False)
```
